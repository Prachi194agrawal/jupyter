How do i apply multiple filter criteria to a panda data frame?

```python
import pandas as pd
```

```python
movies=pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[movies.duration>=200]
```

```python
movies.shape
```

```python
movies[(movies.duration>=200) &(movies.genre=='Drama')]
```

```python
False or False
```

```python
True and False
```

```python
movies[(movies.content_rating=='R') &(movies.genre=='Drama')]
```

```python
(movies.content_rating=='R') &(movies.genre=='Drama')
```

```python
movies[(movies.genre=='Crime')|(movies.genre=='Drama')]
```

```python
movies.genre.isin(['Crime','Drama','Action'])
```

```python
movies[movies.genre.isin(['Crime','Drama','Action'])]
```
